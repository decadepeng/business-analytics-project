```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
color = sns.color_palette()
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
country_wise_df = pd.read_csv('country_wise_latest.csv')
```

```python
country_wise_df.head()
```

```python
country_wise_df_tidy = country_wise_df[['1 week % increase', 'Deaths / 100 Cases', 'Recovered / 100 Cases']]
```

```python
country_wise_df_tidy.head()
```

```python
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=3)
kmeans = k_means.fit(country_wise_df_tidy)
kmeans.cluster_centers_



```

```python
country_wise_df_tidy['cluster'] = kmeans.labels_
country_wise_df_tidy.head(30)
```

```python
type(country_wise_df_tidy['cluster'][0])
```

```python
country_wise_df_tidy['country'] = country_wise_df['Country/Region']
country_wise_df_tidy.head()
```

```python
country_wise_df_tidy
country_wise_3D_plot = px.scatter_3d(
    data_frame = country_wise_df_tidy,
    x = '1 week % increase',
    y = 'Deaths / 100 Cases',
    z = 'Recovered / 100 Cases',
    template = 'ggplot2',
    title = 'Covid Country-wise Analysis',
    color = 'cluster',
    hover_name = 'country',

)
```

```python
country_wise_3D_plot.show()
```
